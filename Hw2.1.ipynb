```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"**

```python
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
birth_data = pd.read_html(url, thousands='.', decimal=',', match='Коефіцієнт народжуваності в регіонах України')[0]
birth_data
```

**Вивести перші рядки таблиці за допомогою методу head**

```python
birth_data.head()
```

**Визначте кількість рядків та стовпців у датафреймі (атрибут shape)**

```python
birth_data.shape
```

**Замініть у таблиці значення "—" на значення NaN**

```python
birth_data.replace('—', np.nan, inplace=True)
birth_data
```

**Визначте типи всіх стовпців за допомогою dataframe.dtypes**

```python
birth_data.dtypes
```

**Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"**

```python
birth_data[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']] = birth_data[['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']].apply(pd.to_numeric)
birth_data.dtypes
```

**Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)**

```python
birth_data.isnull().sum()/28*100
```

**Видаліть з таблиці дані по всій країні, останній рядок таблиці**

```python
birth_data.drop(birth_data.index[-1])
```

**Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)**

```python
birth_data.fillna(birth_data.mean(numeric_only=True)).round(1)
```

**Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні**

```python
birth_data[birth_data['2019'] > birth_data['2019'].mean()][['Регіон','2019']]
```

**У якому регіоні була найвища народжуваність у 2014 році?**

```python
birth_data.loc[birth_data['2014'].idxmax()]['Регіон']
```

**Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році**

```python
width = birth_data.shape[0] * 0.025
plt.bar(birth_data['Регіон'], birth_data['2019'], width=width)
plt.tick_params(labelrotation = 90)
plt.title('Народжуваність по регіонах України у 2019 році')
plt.ylabel('Народжуваність')
plt.show()
```
